```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 100  # Number of consumers
a = 1  # Cost of travel for consumers
c = 1  # Marginal cost of production for firms

# Define the linear market
X = np.linspace(0, 1, N)  # Consumers' locations

# Define the firms' locations
x1 = 1/3  # Firm 1 location
x2 = 2/3  # Firm 2 location

# Define demand functions
def demand_1(x):
    return (a - abs(x - x1)) / 2

def demand_2(x):
    return (a - abs(x - x2)) / 2

# Define profits
def profit_1(x):
    return demand_1(x) * (x - c)

def profit_2(x):
    return demand_2(x) * (1 - x - c)

# Find the Nash equilibrium
def nash_eq():
    x1_new = (a - c) / 2
    x2_new = 1 - x1_new
    return x1_new, x2_new

x1_eq, x2_eq = nash_eq()
print("Nash equilibrium:", x1_eq, x2_eq)

# Plot the demand and profit functions
plt.plot(X, demand_1(X), label="Demand for Firm 1")
plt.plot(X, demand_2(X), label="Demand for Firm 2")
plt.plot(X, profit_1(X), label="Profit for Firm 1")
plt.plot(X, profit_2(X), label="Profit for Firm 2")
plt.axvline(x=x1_eq, linestyle="--", color="gray", label="Nash Equilibrium")
plt.axvline(x=x2_eq, linestyle="--", color="gray")
plt.legend()
plt.show()
```
